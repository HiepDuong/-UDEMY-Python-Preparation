```python
import matplotlib.pyplot as plt

%matplotlib inline
#to see plot inside jupyter notebook
```

```python
import numpy as np
x = np.linspace(0,5,11)
y= x **2
```

# to iterate the axes object use
 - for current_ax in axes.ravel():
    - current_ax.plot(x,y)

```python

#generate 2 rows 2 columns
#this is tuple unpacking
# axes is a 2x2 numpy array where each element is an Axes object.
fig,axes = plt.subplots(nrows =2, ncols = 2)
#avoid overlapping 
plt.tight_layout()

#If you try to iterate over axes directly in a single loop, 
#you're iterating over rows of this array, and each row is itself a numpy array.

#have to use axes.ravel to flatten out
# for current_ax in axes.ravel():
#     current_ax.plot(x,y)

axes[1, 0].plot(y, x)  # For the 2nd row 1st column
axes[1,0].set_title("bottom left plot")
axes[0,1].plot(x,y)
axes[0,1].set_title("top right plot")
```

# Figure Size and DPI

```python
fig = plt.figure(figsize = (8,2)) #increase figure size
ax =fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X lable')
ax.plot(x,x**2)

```

```python
fig, axes = plt.subplots(nrows =2, ncols =2, figsize = (8,2)) #increase figure size
# ax =fig.add_axes([0,0,1,1])
axes[0,0].plot(x,y)
axes[1,1].plot(y,x)
plt.tight_layout()
```

### legends
- label text to clarify what plot is what plot

```python
fig = plt.figure(figsize = (8,2)) #increase figure size
ax =fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X lable')
ax.plot(x,x**2, label ='x squared')
ax.plot(x, x**3, label = 'x cubed')
# usually the legend lable can overlap with line
#to avoid overlap use legend(loc = 0) 0 is best location 

ax.legend(loc = 10 )
```

```python
fig = plt.figure(figsize = (8,2)) #increase figure size
ax =fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X lable')
ax.plot(x,x**2, label ='x squared')
ax.plot(x, x**3, label = 'x cubed')
# usually the legend lable can overlap with line
#to avoid overlap use legend(loc = 0) 0 is best location 

ax.legend(loc = 0 )
```
