```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#to make heatmap work properly, need to be in matrix form
#(index name and column name match up)

#1 corr data
tc =tips.corr()
```

```python
#pass annotation (annot): show numerical value
#cmap

sns.heatmap(tc, annot = True, cmap ='coolwarm')
```

```python
fp = flights.pivot_table(index= 'month', columns= 'year', values = 'passengers')
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor = 'white', linewidth = 1)
```

Clustermap

```python
sns.clustermap(fp, cmap = 'coolwarm', standard_scale =1)
#cluster to most similar month,c olumn and rows that is similar
```
