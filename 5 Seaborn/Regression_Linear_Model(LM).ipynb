```python
#l.M plot: display linear model with seaborn
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#markers: hue with M and F => 2 separtae markers
#scatter_kws = {'s':200} increasing the size of marker
sns.lmplot(x='total_bill', y ='tip', data = tips, hue = 'sex', markers=['o','v'],
           scatter_kws = {'s':200})
```

```python
sns.lmplot(x='total_bill', y ='tip', data = tips,col = 'day', hue = 'sex',aspect = 0.6, height =8)
#separate by collumn col ='sex'
#separate by row  row ='time'
#ratio  between height and grid, =0.6
#use height instead of size()
#chang
```
