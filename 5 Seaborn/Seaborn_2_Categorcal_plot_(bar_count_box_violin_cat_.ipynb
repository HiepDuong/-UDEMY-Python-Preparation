```python
#Categorial Plot
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x= 'sex',y= 'total_bill', data= tips, estimator = np.std)
#show average total_bill
#estimator is a funcion within a function
#geting standard deviation, aggregate visualization of groupby mechanism
```

# Countplot()

```python
sns.countplot(x = 'sex', data =tips)
#bar plot but y axis is count
```

#Whole power of seaborn
- => see split per day and split by smoker
- => allow to add hue indicator to see another layer of d√¢t

```python
sns.boxplot(x='day', y= 'total_bill', data =tips, hue = 'smoker')
#show core tiles of data set
# whisker extend show stdm
# point outside of std is outlier

```

```python
#can use hue, split
sns.violinplot(x='day',y= 'total_bill', data =tips, hue ='sex', split =True)
#combine kernel density plot and box plot => show more informaton
#wdith shwo distribution of data, wider the width = higher density
#can have multiple cateories
# Interquartile Range (IQR):  thicker section in the middle
```

Strip plot

```python
sns.stripplot(x = 'day', y = 'total_bill', data=tips, jitter = True, hue ='sex', dodge = True)
#jitter to see density better
#using dodge to split()
```

```python
#sns.combine violin plot
sns.violinplot(x='day',y='total_bill', data =tips)
sns.swarmplot(x='day',y='total_bill', data =tips, color = 'black')
#no overlap combine scatter and violin plot
#sometime not scale well to number  => alot of computation dont use on large dataset
# in general dont use swarm plot
```

```python
#factoplot() most general form
#(newer version : catplot, older: factorplot)
sns.catplot(x='day', y= 'total_bill', data=tips, kind ='bar')
```
