- ax.plot(x,y, color = '#FF8C00', lw =4, alpha =0.5, ;s ='--', marker = 'o', markersize = 20)
### markerfacecolor
- (, ..., markerfacecolor ='yellow')
### markeredgewidth (edge of the marker)
- (, ..., markeredgewidth = 3)
### markeredgecolor (color of edge)
- (, ..., markerfacecolor ='yellow')

# Special plot type:
- boxplot, histogram, scatter

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#to see plot inside jupyter notebook
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= x **2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'orange')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00', lw =2, alpha =0.5, linestyle ='--', marker = '*', 
        markersize = 20, markerfacecolor = 'yellow', markeredgewidth =3, markeredgecolor = 'green') # RGB Hex Code
```

# Control Over Axis appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C00', lw =2, alpha =0.5, linestyle ='--', marker = '*') # RGB Hex Code
ax.set_xlim([0,1])
#list of lower bound and upper bound
ax.set_ylim([0,2])
```
