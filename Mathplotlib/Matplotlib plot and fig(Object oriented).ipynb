# Object Oriented Method

```python
fig = plt.figure() 
# Creating a figure object an imaginary blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])
# take lsit with left bottom wdith and height arugment from 0 to 1 
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Object Oriented Title")
# create a plot using object oriented method => more control
# similar to plot.plot
```

```python
fig = plt.figure()
axes1 =fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
# from the left
# from the bottom
# width
# and height
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```
